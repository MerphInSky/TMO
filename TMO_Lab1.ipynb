```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
```

В качестве набора данных выбран **Breast Cancer Wisconsin (Diagnostic) Dataset** из библиотеки scikit-learn.

Данный датасет содержит характеристики ядер клеток, извлеченные из изображений тонкоигольных аспиратов (FNA) опухолей молочной железы. Задача — бинарная классификация на доброкачественные (`1`) и злокачественные (`0`) образования.

Из полученных данных видно, что датасет состоит из 569 записей и 31 столбца (30 признаков + целевой). Все признаки имеют числовой тип данных (`float64`), а целевая переменная — `int64`. Как видно из вывода `df.info()`, пропущенные значения отсутствуют.

```python
df.hist(figsize=(18, 15), bins=20, color='skyblue', edgecolor='black')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

Гистограммы показывают распределение каждого признака. Некоторые признаки (например, `mean area`, `perimeter`, `radius`) имеют правостороннее смещение, что типично для физических измерений. Другие распределения близки к нормальному.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='mean area', palette='Set2')
plt.title('Распределение средней площади ядра в зависимости от типа опухоли')
plt.xlabel('Тип опухоли (Target: 0-злокачественная, 1-доброкачественная)')
plt.ylabel('Средняя площадь (mean area)')
plt.show()
```

На графике видно, что у злокачественных опухолей (`0`) медиана и межквартильный диапазон площади ядра значительно выше, чем у доброкачественных (`1`). Это подтверждает, что `mean area` является сильным предиктором для классификации.

```python
# Выбираем 5 наиболее информативных признаков + target для наглядности
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'target']
sns.pairplot(df[selected_features], hue='target', palette='bright', diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(14, 12))
corr_matrix = df.corr()
# annot=False добавлен из-за большого количества признаков (30), чтобы не перегружать график
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5, annot=False, cbar_kws={'shrink': 0.8})
plt.title('Матрица корреляции признаков')
plt.show()
```

Как мы видим на тепловой карте, многие признаки сильно коррелируют друг с другом (например, `mean radius`, `mean perimeter`, `mean area` имеют корреляцию ~0.99, что указывает на мультиколлинеарность).

С целевой переменной `target` сильнее всего коррелируют признаки размера и формы клеток (`mean area`, `mean radius`), в то время как `mean smoothness` и `mean symmetry` демонстрируют слабую линейную связь с целевым классом.
