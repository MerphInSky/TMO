```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Я буду использовать готовый датасет Palmer Penguins

```python
df = sns.load_dataset('penguins')

df.head()
```

```python
df.isnull().sum()
```

Строки с пропусками в колонке `sex` можно удалить, так как их немного.

```python
clear = df.dropna(subset=['sex'])
clear.isnull().sum()
```

Числовые признаки с пропусками мы заменим на медианные значения.

```python
clear['bill_length_mm'] = clear['bill_length_mm'].fillna(clear['bill_length_mm'].median())
clear['bill_depth_mm'] = clear['bill_depth_mm'].fillna(clear['bill_depth_mm'].median())
clear['flipper_length_mm'] = clear['flipper_length_mm'].fillna(clear['flipper_length_mm'].median())
clear['body_mass_g'] = clear['body_mass_g'].fillna(clear['body_mass_g'].median())
clear.isnull().sum()
```

От пропусков избавились. Переходим к кодированию категориальных признаков.

```python
clear.info()
```

Закодируем пол как 0 и 1.

```python
clear['sex'] = clear['sex'].map({'Male': 0, 'Female': 1})
clear.info()
```

Вид пингвина (`species`) мы закодируем от 1 до 3, так как его можно упорядочить.

```python
clear['species'] = clear['species'].map({'Adelie': 1, 'Chinstrap': 2, 'Gentoo': 3}).astype(int)
clear.info()
```

Закодируем one-hot колонку `island`

```python
island_dummies = pd.get_dummies(clear['island'], prefix='island')
island_dummies = island_dummies.astype(int)
clear = pd.concat([clear, island_dummies], axis=1)
clear = clear.drop('island', axis=1)
clear.info()
```

Масштабируем данные.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Масштабируем все признаки, кроме целевого `species`
clear_scaled = pd.DataFrame(scaler.fit_transform(clear.drop('species', axis=1)))

clear_scaled.describe()
```
